```python
import pandas as pd
```

```python
df = pd.read_csv('FaceBook.csv' , delimiter=';')
```

```python
df
```

```python
df.head()
```

```python
##Creating the sub DATASETS
```

```python
df1 = df[['like','share']]
```

```python
df2 = df[['comment','Type']]
```

```python
df1
```

```python
df2
```

```python
##Merge Dataset
```

```python
merged_data = pd.merge(df1,df2,left_on='like', right_on='comment')
```

```python
merged_data
```

```python
merged_data.duplicated().sum()
```

```python
new_dataset = merged_data.drop_duplicates()
```

```python
##Sorting of the dataset
```

```python
new_dataset.sort_values(by=['like'] , ascending=False)
```

```python
##Transpose the dataset
```

```python
new_dataset.transpose()
```
