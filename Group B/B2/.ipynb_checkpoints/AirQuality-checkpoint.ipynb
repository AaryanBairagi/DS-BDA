```python
import pandas as pd
import numpy as np
```

```python
df_air = pd.read_csv('AirQuality.csv',delimiter=';' , decimal=',')
```

```python
df_air.head()
```

```python
##DATA CLEANING
df_air.isnull().sum()
```

```python
df_air_clean = df_air.drop(['Unnamed: 15','Unnamed: 16'] , axis=1)
```

```python
df_air_clean.dropna()
```

```python
##DATA INTEGRATION
sample_df = df_air_clean.sample(n=200 , random_state=42)
```

```python
df1 = sample_df[['CO(GT)','C6H6(GT)','NOx(GT)','NO2(GT)' , 'NMHC(GT)']]
df2 = sample_df[['T','RH','AH','NMHC(GT)']]
```

```python
df1.head()
```

```python
df2.head()
```

```python
merged_data = pd.merge(df2,df1,on='NMHC(GT)',how='inner')
```

```python
merged_data
```

```python
merged_data.duplicated().sum()
```

```python
new_merged_data = merged_data.drop_duplicates(['NMHC(GT)'])
```

```python
# new_merged_data = merged_data.drop_duplicates(['T'])
```

```python
new_merged_data
new_merged_data.info()
```

```python
##Error Correction
```

```python
def remove_outlier(column): 
    column = column.copy()
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3-Q1
    threshold = 1.5*IQR
    outlier_mask = (column<Q1-threshold) | (column>Q3+threshold)
    return column[~outlier_mask]
```

```python
col_name = ['T','RH','AH', 'NMHC(GT)' , 'CO(GT)','C6H6(GT)','NOx(GT)','NO2(GT)']
```

```python
for col in col_name:
    new_merged_data.loc[: , col] = remove_outlier(new_merged_data[col])
```

```python
new_merged_data.reset_index(drop=True , inplace=True)
```

```python
new_merged_data.fillna(new_merged_data.mean(numeric_only=True))
```

```python
##DATA TRANSFORMATION
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression , LogisticRegression
from sklearn.metrics import r2_score , confusion_matrix , accuracy_score , mean_squared_error  , mean_absolute_error

new_data = new_merged_data.dropna()
```

```python
x = new_data.drop(columns=['T'])
y = new_data['T']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
scaler=StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
##DATA MODELING
##1. Linear Regression
linearRegression = LinearRegression()
```

```python
linearRegression.fit(x_train_scaled , y_train)
```

```python
prediction = linearRegression.predict(x_test_scaled)
```

```python
r2 = r2_score(y_test,prediction)
mse = mean_squared_error(y_test,prediction)
mae = mean_absolute_error(y_test,prediction)
print('R2 SCORE: ' , r2)
print('MSE: ' , mse)
print('MAE: ',mae)
# R² = 1 means the model perfectly predicts the data (100% accuracy).
# R² = 0 means the model doesn't explain anything about the data (no better than just guessing the average value).
# R² negative means the model is doing worse than guessing the average value.
```
